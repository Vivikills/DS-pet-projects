```python
import requests
import re
salary = lambda x: [x['salary']['from'],x['salary']['to'],x['salary']['currency'],x['salary']['gross']] if x['salary'] is not None else [None]*4  #те возвращаем ноне 4 раза
snippet = lambda x: [x['snippet']['responsibility'], x['snippet']['requirement']] if x['snippet'] is not None else [None]*2
snippet1 = lambda x: [x['snippet']['responsibility']] if x['snippet'] is not None else [None]*2

vac = [] #список для выкачивания 100 вакансий
vac1 = [] #cписок для выкачивания требований для 100 вакансий

for i in range (10): #хотим 1000 вакансий, но хх выдает по 100 вакансий, значит нам нужно 10 листов с 100 вакансиями
  url = 'https://api.hh.ru/vacancies?text=Программист&per_page=100&area=1&area=2&page={}'.format(str(i))
  vac+=[[x['id'],x['name']]+salary(x)+[x['employer']['name']]+[x['area']['name']]+snippet(x) for x in requests.get(url).json()['items']]
  vac1+=[snippet1(x) for x in requests.get(url).json()['items']]
```

```python
iskl = ['в','на', 'с', 'k', 'и', 'по', 'для', '.', ',', ':', '!', '"', "'", '[', ']', '-', '—', '(', ')']
print(len(vac1))
for i in range (len(vac1)):
  vac1[i] = [word for word in vac1[i] if word is not None]
  vac1[i] = ''.join(vac1[i])
  vac1[i] = vac1[i].lower()
  vac1[i] = re.split('[().,;\s\s+]', vac1[i])
  vac1[i] = [words for words in vac1[i] if words not in iskl and words != '']
  print(vac1[i])
```

```python
_dict = dict()
for j in range (len(vac1)):
  for word in vac1[j]:
      _dict[word] = _dict.get(word, 0) + 1

# сортируем словарь посредством формирования списка (значение, ключ)
_list = []
for key, value in _dict.items():
    _list.append((value, key))
    _list.sort(reverse=True)

# печатаем первые 5 самых используемых слов
a = []
print('Первые 5 самых используемых слов:')
for freq, word in _list[0:5]:
    a.append(word)
    print(f'{word:>5} -> {freq:>3}')
print(a)
```
